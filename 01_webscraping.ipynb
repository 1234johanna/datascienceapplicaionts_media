# Gathering Text on the Web

```python
from requests import get # get: um Sachen von einer Website zu holen = eine Funktion
from bs4 import BeautifulSoup # BeautifulSoup ist eine Klasse
from time import sleep
```

# Requesting & Parasing Web-Content in Python

```python
base_url = "https://quotes.toscrape.com/page/"

corpus = []

for page in range (1, 20): 
    url = base_url + str(page) #um jede Seite aufzurufen

    print("Requesting", url)
    
    response = get(url)#response 200 ist eine gute Response


    soup = BeautifulSoup(
        response.content,
        "html.parser" #html ist auch eine Bibliothek in Python
    )

    quotes = soup.find_all(
        "div", 
        attrs={"class": "quote"}
        )

    for quote  in quotes:   
        textquote = quote.find_next("span", attrs={"class": "text"})
        corpus.append(textquote.text)

    sleep(5)
```
